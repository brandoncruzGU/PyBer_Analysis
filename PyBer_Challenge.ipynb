```python
# Add Matplotlib inline magic command
%matplotlib inline

# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"
```

```python
# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

```python
##Merge the Datasets

# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

```python
## Deliverable 1: Summary DataFrame
```

```python
#Setting the index by city type
city_type = pyber_data_df.set_index(["type"])
city_type
```

```python
#Getting the total number of rides for each city type.
total_rides = city_type.groupby(["type"]).count()["ride_id"]
print(total_rides)
```

```python
#Getting the total drivers for each city type.
total_drivers = city_data_df.groupby("type").sum()["driver_count"]
total_drivers
```

```python
#Getting the total amount of fares for each city type.
total_fare = city_type.groupby(["type"]).sum()["fare"]
total_fare
```

```python
#Getting the average fare per ride for each city type.
average_fare_by_ride = total_fare/total_rides
average_fare_by_ride
```

```python
#Getting the average fare per driver for each city type.
average_fare_by_driver = total_fare/total_drivers
average_fare_by_driver
```

```python
#Creating the PyBer Summary Dataframe
pyber_summary_df = pd.DataFrame(
                    {'Total Number of Rides':total_rides, 
                      'Total Number of Drivers':total_drivers,
                      'Total Amount of Fares':total_fare,
                      'Average Fare per Ride':average_fare_by_ride,
                      'Average Fare per Driver':average_fare_by_driver})
pyber_summary_df
```

```python
#Cleaning the dataframe and deleting the index.
pyber_summary_df.index.name = None

pyber_summary_df
```

```python
#Format the data columns.
pyber_summary_df["Total Amount of Fares"] = pyber_summary_df["Total Amount of Fares"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Ride"] = pyber_summary_df["Average Fare per Ride"].map("${:.2f}".format)
pyber_summary_df["Average Fare per Driver"] = pyber_summary_df["Average Fare per Driver"].map("${:.2f}".format)
pyber_summary_df
```

```python
## Deliverable 2: Create a multiple line plot that shows the total weekly of the fares for each type of city.
```

```python
#Read the merged dataset.
pyber_data_df
```

```python
#Using groupby() to create a new DataFrame showing the sum of the fares for each date where the indices are the city type and date.
pyber_type_df = pyber_data_df.groupby(["type", "date"]).sum(["fare"])
pyber_type_df.tail()
```

```python
#Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
pyber_type_df = pyber_type_df.reset_index()
pyber_type_df.tail()
```

```python
#Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' to get the total fares for each type of city by the date.
pyber_type_df_pivot = pyber_type_df.pivot(index="date", columns="type", values="fare")
pyber_type_df_pivot.head(10)
```

```python
#Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
pyber_type_df_pivot.loc['2019-01-01':'2019-04-29']
```

```python
#Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
pyber_type_df_pivot.index = pd.to_datetime(pyber_type_df_pivot.index)
```

```python
#Check that the datatype for the index is datetime using df.info()
pyber_type_df_pivot.info()
```

```python
#Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
pyber_type_df_week = pyber_type_df_pivot.resample("W").sum()
pyber_type_df_week.head(10)
```

```python
#Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 
pyber_type_df_week.plot(figsize=(25,10))
plt.ylabel('Fare ($USD)')
plt.xlabel('')
plt.title('Total Fare By City Type')
plt.xlim('Jan-01-2019', 'Apr-28-2019')

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')
```
