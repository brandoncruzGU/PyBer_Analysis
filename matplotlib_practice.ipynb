```python
#Import dependencies

import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend.
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, marker="X", color="green", linewidth=2, label='Boston')
# Create labels for the x and y axes.
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
# Set the y limit between 0 and 45.
ax.set_ylim(0, 45)
# Create a title.
ax.set_title("PyBer Fare by Month")
# Add a grid
ax.grid()
# Add the legend.
ax.legend()
```

```python
plt.bar(x_axis, y_axis)
```

```python
# Create a bar chart
plt.bar(x_axis, y_axis, color='green', label='Boston')

# Create a title
plt.title("PyBer Fare By Month")

# Create axis labels
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Create a legend
plt.legend()
```

```python
#Create a horizontal bar chart
plt.barh(x_axis, y_axis, color='magenta', label='Boston')
plt.gca().invert_yaxis()

# Create a title.
plt.title('PyBer Fare By Month')

# Create axis labels
plt.xlabel('Fare($)')
plt.ylabel('Date')

# Create a legend
plt.legend()
```

```python
# Create a bar chart using the object oriented approach
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color = "cyan", label='Chicago')
plt.gca().invert_yaxis()

ax.set_xlabel('Fare($)')
ax.set_ylabel('Date')
ax.set_title('PyBer Fare By Month')
ax.legend()
```

```python
plt.plot(x_axis, y_axis, "o")
```

```python
plt.scatter(y_axis, x_axis, c='red', label="Chicago")
plt.gca().invert_yaxis()
plt.legend()
plt.title('PyBer Fare by Month')
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.xlim(0,50)
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
fig, ax=plt.subplots()
ax.scatter(y_axis, x_axis, label="Boston", s= [i*5 for i in y_axis], c="skyblue", edgecolors="black", alpha=.2)
plt.gca().invert_yaxis()
ax.legend()
ax.set_title("Pyber Fare by Month")
ax.set_xlabel("Fare ($)")
ax.set_ylabel("Date")
ax.set_xlim(0,50)
```

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
plt.subplots(figsize=(8,8))
plt.pie(y_axis, 
        explode=explode_values, 
        colors=colors,
        labels=x_axis, 
        autopct='%.1f%%')

plt.show()
```

```python
explode_value = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
fig, ax=plt.subplots(figsize=(8,8))
ax.pie(y_axis, 
       explode = explode_value,
       labels=x_axis,
       colors = colors,
       autopct= '%1.1f%%',
       shadow=True,
       startangle=-90) 
plt.gca().invert_yaxis()
```

```python
%matplotlib inline

# Import dependencies.
import matplotlib.pyplot as plt
import statistics
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Get the standard deviation of the values of y-axis,
stdev = statistics.stdev(y_axis)
stdev
```

```python
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
plt.bar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
import numpy as np
np.arange(0, 51, step=5.0)
```

```python
from matplotlib.ticker import MultipleLocator
# Increase the size of the plot figure.
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
